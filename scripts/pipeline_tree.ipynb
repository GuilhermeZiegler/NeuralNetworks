```python
import pydot
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo direcionado
G = nx.DiGraph()

# Adicionar nós
G.add_node('Input', label='Input (7 time steps)\n[Feature1, Feature2, ...]')
G.add_node('Feedforward', label='Feedforward Pass')
G.add_node('NeuralNet', label='Neural Network Layers\n(LSTM/GRU, Activation Func.)')
G.add_node('Output', label='Predicted Output\n(Prediction for next time step, e.g., Y8)')
G.add_node('Loss', label='Calculate Loss/Error\n(MSE, BCE, etc.)')
G.add_node('Backprop', label='Backpropagate Error/Gradients\n(via chain rule)')
G.add_node('Update', label='Update Weights\n(via Gradient Descent)')

# Adicionar arestas (conexões)
G.add_edge('Input', 'Feedforward')
G.add_edge('Feedforward', 'NeuralNet')
G.add_edge('NeuralNet', 'Output')
G.add_edge('Output', 'Loss')
G.add_edge('Loss', 'Backprop')
G.add_edge('Backprop', 'Update')
G.add_edge('Update', 'Feedforward')  # Laço de repetição para os próximos Epochs

# Gerar o gráfico com as labels
pos = nx.spring_layout(G, seed=42)  # Layout para posicionamento dos nós
labels = nx.get_node_attributes(G, 'label')

# Criar o gráfico e definir o eixo explicitamente
fig, ax = plt.subplots(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrows=True, ax=ax)
nx.draw_networkx_labels(G, pos, labels=labels, font_size=9, ax=ax)

# Mostrar o gráfico explicitamente
plt.title('Neural Network Training Pipeline', fontsize=14)
plt.show()

```
